```python
class circle:
    pi=3.14
    def __init__(self,radius=1):
        self.radius=radius
        self.area=radius*radius*circle.pi
    def SetRadius(self,new_radius):
     self.radius=new_radius
     self.area=new_radius*new_radius*self.pi
    def getcircumference(self):
        return self.radius*self.pi*2
```

```python
c1 = circle()
print("The radius is:", c1.radius)
print("The area is:", c1.area)
print("The Circumference is:, c1.getcircumference")
```

```python
c1.SetRadius(4)
print("The radius is:", c1.radius)
print("The area is:", c1.area)
print("The Circumference is:, c1.getcircumference")
```

```python
class track:
    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2
        
    def distance(self):
      x1,y1 = self.coor1
      x2,y2 = self.coor2
      return ((x2-x1)**2 + (y2-y1)**2)**0.5
    
    
    def slope(self):
      x1,y1 = self.coor1
      x2,y2 = self.coor2
      return (y2-y1)/(x2-x1)
    
    
```

```python
c1 = (12,9)
c2 = (3,8)
```

```python
find = track(c1,c2)
```

```python
find.distance()
```

```python
find.slope()
```

```python
#bank account detail
class Account:
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance
        
    def __str__(self):#str=string(print the msg)
      return f'Account owner:---------{self.owner}\nAccount balance:--------${self.balance}'
#f is formatting
    def deposit(self,dep_amt):
        self.balance += dep_amt #jitna paise h hmare pass h usme hmne jitna add kiya h vo add hoke show ho
        print("Deposite Accepted")
        
    def withdraw(self,wd_amt):
        if self.balance>= wd_amt:
            self.balance -= wd_amt
            print("withdraw Accepted")
        
        else:
            print("Funds unavailable")
```

```python
acc1 = Account('Raunit',20000)
```
