```python


fig = go.Figure(data=go.Scatter(x=df['Distance'], y=df['DepTime'],mode='markers', marker=dict(color='red')))
fig.update_layout(title='Distance vs Departure Time', xaxis_title='Distance',yaxis_title='DepTime')
# Display the figure
fig.show()
```

```python
line_data = df.groupby('Month')['ArrDelay'].mean().reset_index()
line_data
```

```python
fig = go.Figure(data=go.Scatter(x=line_data['Month'], y =line_data['ArrDelay'], mode='lines', marker=dict(color='green')))
fig.update_layout(title='Month Vs Arrival Delay in timeplot',xaxis_title='Months',yaxis_title='Delay of flight Arrival(in minutes)')
fig.show()
```

```python
bar_data = df.groupby(['DestState'])['Flights'].sum().reset_index()
bar_data
```

```python
fig = pr.bar(bar_data, x="DestState", y="Flights", title='Total no of flights to dest state split by reporting airline')
fig.show()
```

```python
bub_data = df.groupby('Reporting_Airline')['Flights'].sum().reset_index()
bub_data
```

```python
fig_bub = pr.scatter(bub_data, x="Reporting_Airline", y="Flights",title='No of flights per reporting airline', size="Flights",hover_name="Reporting_Airline", size_max=60)
fig_bub.show()
```
